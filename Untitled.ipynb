```python
%matplotlib inline
import csv
import numpy as np
from random import shuffle
def read_csv(fileName):
    allRows = []
    with open(fileName, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            allRows.append(row)
    shuffle(allRows)
    allRows = allRows[:600]
    return np.array(allRows)

def save_edgelist(matrix, fileName):
    fileName = '/home/kujain/node2vec/graph/'+fileName+'.edgelist'
    fileToWrite = open(fileName, 'w')
    writer = csv.writer(fileToWrite,delimiter=' ')
    for i in range(matrix.shape[0]):
        for j in range(i,matrix.shape[1]):
            if(matrix[i,j] > 1e-8):
                writer.writerow([i,j,matrix[i,j]])
def read_emb(fileName):
    fileName = '/home/kujain/node2vec/emb/'+fileName+'.emd'
    allRows = []
    with open(fileName, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            allRows.append(row[0].split(' '))
    allRows = allRows[1:]
    D = len(allRows[0])
    N = len(allRows)
    M = np.zeros((N,D-1))
    for row in allRows:
        #print(row)
        M[int(row[0]),:] = np.array([float(num) for num in row[1:]])
    return M
```

```python
def getXY(data):
    return data[:,:-1].astype(np.float),data[:,-1].astype(np.int)
```

```python
from sklearn.neighbors import kneighbors_graph
import scipy
def getGraphMatrix(X,typeOf,param):
    if(typeOf == 'n nearest'):
        graph = kneighbors_graph(X, param, mode='connectivity', include_self=False)
        return graph
    if(typeOf == 'gaussian'):
        from scipy.spatial.distance import pdist, squareform
        knGraph = kneighbors_graph(X, param, mode='connectivity', include_self=False)
        pairwise_dists = squareform(pdist(X, 'euclidean'))
        K = scipy.exp(-pairwise_dists ** 2 / 1 ** 2)
        #return knGraph.toarray()
        return K*(knGraph.toarray())
    if(typeOf == 'epsilon'):
        from scipy.spatial.distance import pdist, squareform
        pairwise_dists = squareform(pdist(X, 'euclidean'))
        return pairwise_dists
    return X
```

```python
def simpleKMeans(data,clusters,neigh,graph='gaussian'):
    #data = read_csv('./datasets/'+dataset+'.csv')
    from sklearn.cluster import KMeans
    X,Y = getXY(data)
    kmeans = KMeans(n_clusters=clusters,random_state=0).fit(X)
    labels = kmeans.labels_
    return X,Y,labels,None
```

```python
import numpy as np
import random

def get_initial_prototypes(X, k, seed):
    random.seed(seed)
    np.random.seed(seed)
    number_of_samples = X.shape[0]
    all_ids = [i for i in range(X.shape[0])]
    random.shuffle(all_ids)
    prototypes = all_ids[:k]
    return prototypes

def get_clusters(X, prototypes):
    X_selective = X[:,prototypes]
    closest_cluster_ids = np.argmin(X_selective, axis=1)
    clusters = {}
    for i in range(len(prototypes)):
        clusters[i] = [prototypes[i]]
    for i, cluster_id in enumerate(closest_cluster_ids):
        if(i not in prototypes):
            clusters[cluster_id].append(i)
    return clusters

def update_prototypes(X, clusters):
    prototypes = [0 for i in range(len(clusters))]
    total_cost = 0.0
    labels = [0 for i in range(X.shape[0])]
    for i in range(len(clusters)):
        cluster = clusters[i]
        cluster.sort()
        #X_selective = X[cluster, cluster]
        X_selective = X[np.ix_(cluster,cluster)]
        distance_sum = np.sum(X_selective, axis = 1)
        new_prototype = np.argmin(distance_sum)
        cur_cost = np.min(distance_sum)
        total_cost += cur_cost
        prototypes[i] = cluster[new_prototype]
        for c in cluster:
            labels[c] = i
    return prototypes, total_cost, labels

def perform_k_means_algorithm(X, k, movement_threshold_delta=0,seed=0):
    new_prototypes = get_initial_prototypes(X, k, seed)
    should_terminate = False
    new_cost = np.sum(X)
    labels = []
    iterations = 0
    while not should_terminate:
        previous_prototypes = new_prototypes
        previous_cost = new_cost
        clusters = get_clusters(X, previous_prototypes)
        new_prototypes, new_cost, labels = update_prototypes(X, clusters)
        should_terminate = (previous_cost - new_cost < movement_threshold_delta)
        iterations += 1
    return new_prototypes, labels
```

```python
def unnormSpectral(data,clusters,neigh,graph='gaussian'):
    import numpy.linalg as linalg
    from sklearn.cluster import KMeans
    #data = read_csv('./datasets/'+dataset+'.csv')
    X,Y = getXY(data)
    W = getGraphMatrix(X,graph,neigh)
    W = np.minimum( W, W.transpose() )
    D = np.diag(np.sum(W,axis=1))
    L = D - W
    eigenValues, eigenVectors = linalg.eigh(L)
    eigenValues = eigenValues[:clusters]
    eigenVectors = eigenVectors[:,:clusters]
    kmeans = KMeans(n_clusters=clusters,random_state=0).fit(eigenVectors)
    labels = kmeans.labels_
    return X,Y,labels,W

def normSpectralRW(data,clusters,neigh,graph='gaussian'):
    import numpy.linalg as linalg
    from sklearn.cluster import KMeans
    #data = read_csv('./datasets/'+dataset+'.csv')
    X,Y = getXY(data)
    W = getGraphMatrix(X,graph,neigh)
    W = np.minimum( W, W.transpose() )
    D = np.diag(np.sum(W,axis=1))
    L = D-W
#     Lrw = np.matmul(scipy.sparse.linalg.inv(D),L)
    eigenValues, eigenVectors = linalg.eigh(L-D)
    eigenValues = eigenValues[:clusters]
    eigenVectors = eigenVectors[:,:clusters]
    kmeans = KMeans(n_clusters=clusters,random_state=0).fit(eigenVectors)
    labels = kmeans.labels_
    return X,Y,labels,W

def normSpectralSym(data,clusters,neigh,graph='gaussian'):
    import numpy.linalg as linalg
    from sklearn.cluster import KMeans
    import sklearn
    #data = read_csv('./datasets/'+dataset+'.csv')
    X,Y = getXY(data)
    W = getGraphMatrix(X,graph,neigh)
    W = np.minimum( W, W.transpose() )
    D = np.diag(np.sum(W,axis=1))
    D_sqrt_inv = np.diag(np.reciprocal(np.sqrt(np.sum(W,axis=1))))
    L = D-W
    #D_sqrt_inv = np.linalg.inv(np.sqrt(D))
    Lsym = (D_sqrt_inv @ L) @ D_sqrt_inv
    eigenValues, eigenVectors = linalg.eigh(Lsym)
    eigenValues = eigenValues[:clusters]
    eigenVectors = eigenVectors[:,:clusters]
    eigenVectors = sklearn.preprocessing.normalize(eigenVectors, norm="l2")
    kmeans = KMeans(n_clusters=clusters,random_state=0).fit(eigenVectors)
    labels = kmeans.labels_
    return X,Y,labels,W
```

```python
def ECT(data,clusters,neigh,graph='gaussian',seed=0):
    import numpy.linalg as linalg
    from sklearn.cluster import KMeans
    from scipy.spatial.distance import pdist, squareform
    #data = read_csv('./datasets/'+dataset+'.csv')
    X,Y = getXY(data)
    knGraph = kneighbors_graph(X, neigh, mode='connectivity', include_self=False)
    pairwise_dists = squareform(pdist(X, 'euclidean'))
    from scipy.sparse.csgraph import minimum_spanning_tree
    mst = minimum_spanning_tree(pairwise_dists)
    mst[mst > 0.0] = 1.0
    knGraph = knGraph.toarray()
    mst = mst.toarray()
    knGraph = np.maximum(knGraph,knGraph.T)
    knGraph = np.maximum(knGraph,mst)
    W = scipy.exp(-pairwise_dists ** 2 / 1 ** 2)*knGraph
    W = pairwise_dists*knGraph
    W = np.minimum( W, W.transpose() )
    D = np.diag(np.sum(W,axis=1))
    L = D-W
    Lp = np.linalg.pinv(L)
    #print(Lp)
    Lpd = np.diagonal(Lp)
    secondTerm = Lp
    thirdTerm = Lp.transpose()
    fourthTerm = np.tile(Lpd,(Lpd.shape[0],1))
    firstTerm = fourthTerm.transpose()
    
    N = np.sum(W)*(firstTerm - secondTerm - thirdTerm + fourthTerm)
    #from k_means_proto import perform_k_means_algorithm
    prototypes, labels = perform_k_means_algorithm(N,clusters,1e-6,seed)
    return X,Y,labels,prototypes,W
```

```python
def draw(X, labels, clusters):
    import seaborn as sns
    import matplotlib.pyplot as plt
    sns.set_style('darkgrid')
    sns.set_palette('muted')
    sns.set_context("notebook", font_scale=1.5,
                rc={"lines.linewidth": 2.5})
    palette = np.array(sns.color_palette("hls", clusters))
    plt.scatter(X[:,0], X[:,1], lw=0, s=40, c=palette[labels.astype(np.int)])
    plt.show()

def scatter(M, labels, clusters):
    # We choose a color palette with seaborn.
    from sklearn.manifold import TSNE
    import seaborn as sns
    x = TSNE(random_state=20150101).fit_transform(M)
    palette = np.array(sns.color_palette("hls", clusters))

    # We create a scatter plot.
    #f = plt.figure(figsize=figSize)
    import matplotlib.pyplot as plt
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40,
                    c=palette[labels.astype(np.int)])
    plt.xlim(-25, 25)
    plt.ylim(-25, 25)
    #plt.title(plotName)
    #ax.legend([plotName])
    ax.axis('off')
    ax.axis('tight')
    plt.show()

def simpleStats(trueLabels, predictedLabels):
    import math
    trueLabelsDict = {}
    predictedLabelsDict = {}
    for i in range(len(trueLabels)):
        if(trueLabels[i] not in trueLabelsDict):
            trueLabelsDict[trueLabels[i]] = []
        trueLabelsDict[trueLabels[i]].append(i)
    for i in range(len(predictedLabels)):
        if(predictedLabels[i] not in predictedLabelsDict):
            predictedLabelsDict[predictedLabels[i]] = []
        predictedLabelsDict[predictedLabels[i]].append(i)
    trueLabelsSummary = []
    predictedLabelsSummary = []
    for t in trueLabelsDict:
        trueLabelsSummary.append(trueLabelsDict[t])
    for p in predictedLabelsDict:
        predictedLabelsSummary.append(predictedLabelsDict[p])
    commonElems = np.zeros((len(trueLabelsSummary), len(predictedLabelsSummary)))
    for i in range(commonElems.shape[0]):
        for j in range(commonElems.shape[1]):
            commonElems[i,j] = len(set(trueLabelsSummary[i]).intersection(set(predictedLabelsSummary[j])))
    #print(commonElems)
    nc2 = np.vectorize(lambda x: (x*(x-1))/2)
    a = np.sum(nc2(commonElems))
    common_rows_summed = np.sum(commonElems, 1)
    common_columns_summed = np.sum(commonElems, 0)
    b = np.sum(nc2(common_rows_summed)) - a
    c = np.sum(nc2(common_columns_summed)) - a
    jaccard = (a/(a+b+c))
    fm = a/(math.sqrt((a+b)*(a+c)))
    d = (np.sum(commonElems)*(np.sum(commonElems) - 1))/2
    ari = (a-(((b+a)*(c+a))/d))/((0.5*(b+c+2*a))-(((b+a)*(c+a))/d))
    return [jaccard, fm, ari]

def graphStats(graph,labels):
    labelsDict = {}
    import math
    for i in range(len(labels)):
        if(labels[i] not in labelsDict):
            labelsDict[labels[i]] = []
        labelsDict[labels[i]].append(i)
    labelsSummary = []
    for t in labelsDict:
        labelsSummary.append(labelsDict[t])
    modularity = 0
    totalEdges = (graph > 0).sum()
    coverage = 0
    for i in range(len(labelsSummary)):
        curCluster = list(labelsSummary[i])
        graphCur = graph[np.ix_(curCluster,curCluster)]
        e = (graphCur > 0).sum()
        modularity += float(e)/totalEdges
        coverage += np.sum(graphCur)
        for j in range(len(labelsSummary)):
            if(j != i):
                tempCluster = list(labelsSummary[j])
                graphCur = graph[np.ix_(curCluster,tempCluster)]
                a = (graphCur > 0).sum()/2
                modularity -= math.pow((float(a+e))/totalEdges,2)
    coverage /= np.sum(graph)
    return [modularity,coverage]
                
                
                
                
```
